### 1. What are the most common crop types in the dataset, and how do they correlate with disease severity?

```python
#Find the most common crop types

crop_counts=data['Crop_Type'].value_counts()
print("Most common crop types:\n", crop_counts)
```

```python
#Convert categorical crop types to numeric

data_encoded=pd.get_dummies(data[['Crop_Type','Disease_Severity']])
data_encoded
```

```python
#Calculate correlation with disease severity

correlation_matrix = data_encoded.corr()
correlation_matrix
```

```python
#Extract correlation of crop type with disease severity

crop_severity_corr = correlation_matrix['Disease_Severity']

print("\nCorrelation of crop types with disease severity:\n", crop_severity_corr)
```

### 2. How  does soil type impact the growth rate of plants? Analyze the correlation.

```python
# Convert categorical soil types to numeric using one-hot encoding

data_encoded_1=pd.get_dummies(data[['Soil_Type','Growth_Rate']])
data_encoded_1
```

```python
# Calculate correlation between soil types and growth rate 

correlation_matrix_1=data_encoded_1.corr()
correlation_matrix_1
```

```python
# Extract the correlation with soil types and growth rate

soil_growth_corr=correlation_matrix_1['Growth_Rate']
print("Correlation between soil types and plant growth rate:\n", soil_growth_corr)
```

### 3. Examine the relationship between the region and nutrient deficiency. Which regions have the highest deficiencies?

```python
# Create a contingency table to count occurrences of nutrient deficiencies by region

contigency_table=pd.crosstab(data['Region'],data['Nutrient_Deficiency'])
contigency_table
```

```python
# Add a column for the total deficiencies per region

contigency_table['Total_Deficiencies']=contigency_table.sum(axis=1)
contigency_table
```

```python
# Sort the table by the total deficiencies 
sorted_deficiency=contigency_table.sort_values(by='Total_Deficiencies',ascending=False)
sorted_deficiency
```

```python
# plotting the deficiency by region
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,4))
sns.barplot(data=sorted_deficiency,x='Region',y='Total_Deficiencies',palette='viridis',width=0.5)
plt.title('Nutrient Deficiency by the Region')
plt.xlabel('Nutrient Deficiency')
plt.ylabel('Region')
plt.show()
```
